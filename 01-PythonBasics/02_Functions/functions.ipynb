## Functions
- A function is a block of code that performs a specific task.
- Functions help in organizing code, making it reusable and easier to read.

```python
# why functions?
num = 24
if num % 2 == 0:
  print(f"{num} is even")
else:
  print(f"{num} is odd")
```

```python
def check_even_odd(number):
  if number % 2 == 0:
    return f"{number} is even"
  else:
    return f"{number} is odd"

result = check_even_odd(24)
print(result)
```

```python
# Variable Lengths Arguments
# Positional and Keywords arguments

# Positional Arguments
def print_numbers(*args):
  for numbers in args:
    print(numbers)
```

```python
print_numbers(1, 2, 3, 4, 5, "Ayush")
```

```python
# Keywords Arguments

def print_details(**kwargs):
  for key, value in kwargs.items():
    print(f"{key}:{value}")
```

```python
print_details(name="Krish", age="32", country="India")
```

```python
def print_detail(*args, **kwargs):
  for val in args:
    print(f"Positional args: {val}")
  for key, value in kwargs.items():
    print(f"{key}:{value}")
```

```python
print_detail(1, 2, 3, 4, "Ayush", name="Shivam", age=21, country="India")
```
